```python
import ast

def convert(text):
    try :
        return ast.literal_eval(text)
    except:
        return []
    
```

```python
type(movies['genres'][0])
```

```python
movies['genres'] = movies['genres'].apply(convert)
movies['keywords'] = movies['keywords'].apply(convert)
movies['crew'] = movies['crew'].apply(convert)
movies['cast'] = movies['cast'].apply(convert)
```

```python
def helper(random):
    L = []
    for i in random:
        L.append(i['name'])
        return L
    


def conver3(obj):
    L = []
    counter = 0
    for i in obj:
        if counter != 3:
            L.append(i['name'])
            counter +=1
        else:
            break
        return L
    

def fetch_director(jb):
    L = []
    for i in jb:
        if i['job'] == 'Director':
            L.append(i['name'])
            break
        return L
    
```

```python
movies['genres'] = movies['genres'].apply(convert).apply(helper)
movies['keywords'] = movies['keywords'].apply(convert).apply(helper)
movies['cast'] = movies['cast'].apply(convert).apply(conver3)
movies['crew'] = movies['crew'].apply(convert).apply(fetch_director)
```
