```python
import pandas as pd
import numpy as np


movies = pd.read_csv('tmdb_5000_movies.csv')
credits = pd.read_csv('tmdb_5000_credits.csv')

movies.head(2)
```

```python
movies = movies.merge(credits, on='title')
```

```python
movies.head(2)
```

```python
movies = movies[['movie_id', 'title', 'overview', 'cast', 'crew', 'genres', 'keywords']]
```

```python
movies.head(2)
```

```python
movies['movie_id'].value_counts()
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
import ast

def convert(obj):
    L = []
    # ast.literal_eval string ko list/dictionary mein badal deta hai
    for i in ast.literal_eval(obj):
        L.append(i['name']) 
    return L
```

```python
movies['genres'] = movies['genres'].apply(convert)
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
def convert3(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 3:
            L.append(i['name'])
            counter += 1
        else:
            break
    return L

movies['cast'] = movies['cast'].apply(convert3)
```

```python
def fetch_director(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L

movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies['overview'] = movies['overview'].apply(lambda x: x.split())
```

```python
movies.head(2)
```

```python
movies.isnull().sum()
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies.head(2)
```

```python
movies.tags[0]
```

```python
new_movies = movies.drop(columns=['keywords', 'cast', 'crew', 'overview', 'genres'])
```

```python
new_movies.head(2)
```

```python
new_movies.iloc[0].tags
```

```python
# Change every word of list in paragraph
new_movies['tags'] = new_movies['tags'].apply(lambda x: " ".join(x))
```

```python
new_movies.iloc[0].tags
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()

def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)

new_movies['tags'] = new_movies['tags'].apply(stem)
```

```python
pd.set_option('display.max_colwidth', None)
new_movies['tags'].head(2)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words='english')

# Tags ko 5000 numbers ki list mein badalna
vectors = cv.fit_transform(new_movies['tags']).toarray()
```
